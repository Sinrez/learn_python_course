```python
import openpyxl

streets = {}

wb_obj = openpyxl.load_workbook('/content/data-398-2023-02-27.xlsx') 
sheet_obj = wb_obj.active 

cell_obj = sheet_obj.cell(row = 1, column = 1) 
print(cell_obj.value)

row = sheet_obj.max_row
column = sheet_obj.max_column
  
print("Total Rows:", row)
print("Total Columns:", column)

cell_obj = sheet_obj['A1': 'B6']
  
# Print value of cell object 
# using the value attribute 
for cell1, cell2 in cell_obj:
    print(cell1.value, cell2.value)
```

```python
  # print(cell_obj)
for cell1, cell2  in cell_obj:
        if cell2.value in streets:
            streets[cell2.value] += 1
        else:
            streets[cell2.value] = 1
print(streets)
```

```python
# https://wits.worldbank.org/CountryProfile/en/Country/RUS/Year/2016/TradeFlow/Export

export_2015 = pd.read_excel('/content/WITS-Partner.xlsx', sheet_name=1)
export_2016 = pd.read_excel('/content/WITS-Partner 2016.xlsx', sheet_name=1)
```

```python
export = pd.concat([export_2015, export_2016])
export.sample(10)
```

Сводная таблица позволяет посмотреть итоги (суммы, средние и т.д.) по данным исходной таблицы. В методе pivot_table параметр index - столбцы по которым делается группировка, values - столбец с данными над которыми мы проведем операцию и aggfunc - то как считаем результат:

```python
export.pivot_table(index=['Year'],
    values='Export (US$ Thousand)', aggfunc='sum')
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
export.pivot_table(index=['Year', 'Partner Name'],
    values='Export (US$ Thousand)', aggfunc='sum')
```

```python
export.pivot_table(index=['Year', 'Partner Name'],
    values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values('Export (US$ Thousand)', ascending=False)
```

```python
export_filtered = export[export['Partner Name'] != 'Europe & Central Asia']

export_filtered.pivot_table(index=['Year', 'Partner Name'],
    values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values('Export (US$ Thousand)', ascending=False)
```

```python
exclude = ['Europe & Central Asia', 'East Asia & Pacific', ' World',
    'Unspecified', 'Middle East & North Africa']
export_filtered = export[-export['Partner Name'].isin(exclude)]
#- который обозначает не и метод isin который обозначает входит в список

export_filtered.pivot_table(index=['Year', 'Partner Name'], 
                values='Export (US$ Thousand)', aggfunc='sum')\
            .sort_values('Export (US$ Thousand)', ascending=False)
```

```python
export_pivot = export_filtered.pivot_table(index=['Year', 'Partner Name'],
                values='Export (US$ Thousand)', aggfunc='sum')\
            .sort_values(
                ['Year', 'Export (US$ Thousand)'],
                ascending=[True, False]
            )
export_pivot
```

```python
export_pivot.index.get_level_values('Year') == 2015
```

```python
export_pivot[export_pivot.index.get_level_values('Year') == 2015]
```
